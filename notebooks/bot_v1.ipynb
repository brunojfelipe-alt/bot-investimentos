```python
# ============================================================

# PROJETO: BOT DE ANÁLISE DO MERCADO DE AÇÕES BRASILEIRO

# Autor: Bruno Januário Felipe

# Curso: Matemática Aplicada a Negócios

# Universidade: USP Ribeirão Preto

#

# Objetivo do projeto:

# Desenvolver um sistema automatizado capaz de coletar dados

# do mercado financeiro brasileiro, analisar ações utilizando

# indicadores técnicos e informações de notícias financeiras,

# gerar scores de investimento e avaliar estratégias através

# de backtesting.

#

# O projeto está dividido em versões progressivas (V1 -> V9),

# onde cada versão adiciona novas capacidades analíticas.

# ============================================================

# ============================================================

# V1 — COLETA E VISUALIZAÇÃO DE DADOS

# ============================================================

#

# Objetivo:

# Construir a base do sistema coletando dados históricos de

# ações brasileiras e exibindo informações básicas.

#

# Funcionalidades:

#

# - Baixar dados históricos de ações (ex: PETR4, VALE3)

# - Mostrar preço atual

# - Mostrar variação percentual diária

# - Plotar gráfico de preço ao longo do tempo

# - Criar Modos de Visualização: 
#   1 — Ranking de ações
#   2 — Análise de uma ação
#   3 — Backtest
#   4 — Portfólio do usuário
#   5 — Alertas
#   6 — Notícias

# Conceitos:


# PREÇO ATUAL

# Último preço negociado da ação.

#

# VARIAÇÃO PERCENTUAL

# Mede quanto o preço mudou em relação ao dia anterior.

#

# Fórmula:

#

# variação = (preço_atual - preço_anterior) / preço_anterior

#

# RETORNO PERCENTUAL

# Mede o ganho ou perda em determinado período.

#

# retorno = (preço_final - preço_inicial) / preço_inicial

#

# RETORNO ACUMULADO

# Mede o retorno total ao longo de vários períodos.

#

# Exemplo:

# Se uma ação subiu:

# 5% em janeiro

# 3% em fevereiro

# 2% em março

#

# O retorno acumulado é a combinação desses retornos.

#

# Bibliotecas típicas usadas:

# pandas, yfinance, matplotlib

# ============================================================

# V1.5 — ESTRUTURA DE BANCO DE DADOS

# ============================================================

#

# Objetivo:

# Criar um banco de dados local para armazenar informações

# coletadas pelo sistema.

#

# Dados armazenados:

#

# - preços históricos

# - volume de negociação

# - indicadores técnicos

# - scores gerados pelo sistema

# - notícias financeiras

#

# Vantagens:

#

# - evitar baixar os dados toda vez

# - permitir análises históricas

# - acelerar processamento

#

# Tecnologias possíveis:

#

# SQLite

# CSV

# Parquet

#

# Ideal para projetos iniciais: SQLite ou Parquet

# ============================================================

# V2 — RANKING DE DESEMPENHO

# ============================================================

#

# Objetivo:

# Comparar o desempenho das ações e criar rankings.

#

# Funcionalidades:

#

# - ranking das ações que mais subiram

# - ranking das ações que mais caíram

# - ranking por período:

# - diário

# - semanal

# - mensal

#

# Exemplo:

#

# Top ganhos do dia

#

# 1. WEGE3 +4.2%

# 2. PETR4 +3.1%

# 3. VALE3 +2.7%

#

# Isso ajuda a identificar:

#

# - ações em tendência

# - movimentos fortes do mercado

# ============================================================

# V3 — INDICADORES TÉCNICOS

# ============================================================

#

# Objetivo:

# Adicionar métricas matemáticas usadas em análise técnica.

#

# Indicadores implementados:

#

# MÉDIA MÓVEL (Moving Average)

#

# Média do preço em um número de dias.

#

# Exemplo:

# Média móvel de 20 dias (MM20)

#

# Ajuda a identificar tendência do mercado.

#

# Se preço > média móvel:

# tendência de alta

#

# Se preço < média móvel:

# tendência de baixa

#

#

# VOLUME

#

# Quantidade de ações negociadas em determinado período.

#

# Volume alto pode indicar:

#

# - grande interesse do mercado

# - possíveis movimentos fortes

#

#

# VOLATILIDADE

#

# Mede o quanto o preço oscila ao longo do tempo.

#

# Geralmente calculada usando:

#

# desvio padrão dos retornos

#

# Alta volatilidade:

# mercado instável

#

# Baixa volatilidade:

# mercado mais estável

#

#

# RSI (Relative Strength Index)

#

# Indicador de momentum que mede velocidade das variações

# de preço.

#

# Escala:

#

# 0 → 100

#

# RSI > 70

# ação sobrecomprada (pode cair)

#

# RSI < 30

# ação sobrevendida (pode subir)

#

#

# MACD (Moving Average Convergence Divergence)

#

# Indicador que mede relação entre duas médias móveis.

#

# Usado para detectar:

#

# - mudanças de tendência

# - força do movimento

#

# O MACD analisa a diferença entre:

#

# média móvel curta

# média móvel longa

#

# Quando o MACD cruza sua linha de sinal,

# pode indicar compra ou venda.

# ============================================================

# V3.5 — CORRELAÇÃO ENTRE AÇÕES

# ============================================================

#

# Objetivo:

# Medir relação entre os movimentos de diferentes ações.

#

# Correlação varia entre:

#

# -1  → movimento oposto

# 0  → sem relação

# +1  → movimento igual

#

# Exemplo:

#

# ITUB4 e BBDC4

# geralmente possuem alta correlação

#

# Isso ajuda em:

#

# - diversificação

# - construção de portfólio

# ============================================================

# V4 — SISTEMA DE ALERTAS

# ============================================================

#

# Objetivo:

# Detectar eventos importantes automaticamente.

#

# Exemplos de alertas:

#

# - ação subiu mais de 5%

# - ação caiu mais de 5%

# - volume anormal

# - cruzamento de médias móveis

#

# Inicialmente os alertas podem aparecer:

#

# - no terminal

# - em logs

#

# Futuro:

#

# - notificações via Telegram

# - notificações no celular

# ============================================================

# V5 — COLETA DE NOTÍCIAS FINANCEIRAS

# ============================================================

#

# Objetivo:

# Coletar automaticamente notícias relacionadas ao mercado.

#

# Técnicas utilizadas:

#

# Web Scraping

#

# Possíveis fontes:

#

# - portais financeiros

# - notícias econômicas

# - comunicados de empresas

#

# As notícias serão armazenadas em banco de dados

# para análises futuras.

# ============================================================

# V6 — ANÁLISE DE SENTIMENTO

# ============================================================

#

# Objetivo:

# Usar inteligência artificial para classificar notícias.

#

# Classificação:

#

# POSITIVA

# NEUTRA

# NEGATIVA

#

# Exemplo:

#

# "Empresa anuncia aumento de lucro"

# → sentimento positivo

#

# "Empresa anuncia prejuízo"

# → sentimento negativo

#

# Isso permite avaliar impacto de notícias

# sobre o mercado.

# ============================================================

# V7 — SISTEMA DE SCORE DAS AÇÕES

# ============================================================

#

# Objetivo:

# Criar um score quantitativo para cada ação.

#

# O score combina:

#

# - indicadores técnicos

# - notícias

# - histórico de preços

# - volatilidade

#

# Exemplo de escala:

#

# -100 → extremamente negativo

# 0 → neutro

# +100 → extremamente positivo

#

# Exemplo:

#

# PETR4 → score 72

# VALE3 → score 55

# ITUB4 → score 81

#

# Isso gera um ranking inteligente

# de oportunidades de investimento.

# ============================================================

# V7.5 — OTIMIZAÇÃO DE PORTFÓLIO

# ============================================================

#

# Objetivo:

# Construir portfólios ideais de investimento.

#

# Baseado na teoria moderna de portfólio.

#

# Ideia central:

#

# maximizar retorno

# minimizar risco

#

# usando:

#

# - retornos esperados

# - volatilidade

# - correlação entre ativos

# ============================================================

# V8 — BACKTESTING

# ============================================================

#

# Objetivo:

# Testar estratégias utilizando dados históricos.

#

# Exemplo:

#

# Estratégia:

#

# comprar ações com score > 70

#

# O sistema simula:

#

# - compras

# - vendas

# - retorno acumulado

#

# Isso permite medir:

#

# - retorno da estratégia

# - risco

# - precisão do modelo

# ============================================================

# V9 — INTERFACE GRÁFICA

# ============================================================

#

# Objetivo:

# Criar um dashboard visual do mercado.

#

# Funcionalidades:

#

# - lista de ações

# - gráficos interativos

# - indicadores técnicos

# - scores

# - rankings

# - alertas

#

# O usuário poderá clicar em uma ação

# e visualizar todas as informações

# em uma interface visual.

# ============================================================

# ============================================================

# BIBLIOTECAS PRINCIPAIS DO PROJETO

# ============================================================

# Manipulação de dados

#import pandas as pd
#import numpy as np

# Coleta de dados de mercado

#import yfinance as yf

# Visualização de gráficos

#import matplotlib.pyplot as plt
#import seaborn as sns
#import plotly.express as px

# Indicadores técnicos

#import ta

# Manipulação de datas

#import datetime as dt

# Web scraping

#import requests
#from bs4 import BeautifulSoup

# Processamento de texto (NLP)

#import nltk
#from textblob import TextBlob

# Machine Learning

#from sklearn.model_selection import train_test_split
#from sklearn.ensemble import RandomForestClassifier

# Banco de dados

#import sqlite3

# Sistema operacional

#import os

# Estatística

#from scipy import stats

# Avisos

#import warnings
#warnings.filterwarnings("ignore")
```
